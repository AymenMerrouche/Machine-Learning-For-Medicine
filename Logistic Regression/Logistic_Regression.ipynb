> The classifier is linear, meaning that the decision boundary is hyperplane. An obvious limitation for this classifier is non linearly separable data. In our case, the more the data are linearly separable, the better this classifier performs. For the moons dataset for example, the classifier is, by design, unable to correctly classify all the data. Theoretically, in logistic regression, the assumption is not made on the data distribution, but on the Bayes classifier : it is assumed to be linear, hence the decision boundary is a hyperplane.

# 2 - From scratch binary logistic regression :
> In this section we provide a from scratch implementation of a logistic regression classifier using the Newton-Raphson optimisation method. We will create a class inheriting from sklearn, so that we can use all sklearn's scoring methods, in particular cross validation score.

```python
from sklearn.base import BaseEstimator
from sklearn.discriminant_analysis import LinearClassifierMixin
from numpy.linalg import pinv
class LogReg(BaseEstimator, LinearClassifierMixin):
    """
        Newton Raphson opitimisation to fit a logistic regression binar classifier
    """
    def __init__(self, epochs = 10, bias=True):
        """
            :param bias: boolean, wether to use a bias or not (W.x + b*)
            :param epochs: number of iterations to perform (number of optimisation steps)
        """
        self.bias = bias
        self.epochs = epochs
        
    def fit(self, X, y):
        """
            :param X: data matrix
            :param y: labels
        """
        # convert inputs to numpy arrays
        X = np.asarray(X)
        y = np.asarray(y)
        # initilisation
        if self.bias :
            w = np.zeros(X.shape[1] + 1)
            # add a dimension of 1 in the data matrix to account for the bias : [X, 1]*[W, b] = X*W + 1*b
            X = np.insert(X, X.shape[1], 1, axis=1)
        else:
            w = np.zeros(X.shape[1])

        # training loop
        for epoch in range(self.epochs):
            # gradient
            G = self.first_derivative(w, X, y)
            # hessian
            H = self.second_derivative(w, X)
            # update
            w = w - pinv(H) @ G
        # retrieve intercept and hyperplan
        self.intercept_ = w[-1]
        self.coef_ = w[0:len(w)-1]
        return self

    def decision_function(self, X):
        return X.dot(self.coef_) + self.intercept_

    def predict(self, X):
        # identical to testing probabilities
        # exp(theta.T@X) - 1 >= 0 <=> exp(theta.T@X) >= 1 <=> apply log
        return (self.decision_function(X) > 0)*1
    
    def first_derivative(self,w, X, y):
        """
        First derivative of logreg loss wrt w
        """
        # compute probabilities and error
        probs = self.prob(w, X)
        diff = (y - probs)
        # boradcast shape of diff to multiply (Hadamard multiplicatin)
        diff = np.repeat(np.expand_dims(diff, axis=1), X.shape[1], axis=1)
        # compute sum
        return -np.sum(np.multiply(X, diff), axis = 0)

    def second_derivative(self,w, X):
        """
        Second derivative of logreg loss wrt w
        """
        # compute probabilities
        probs = self.prob(w, X)
        probs = probs * (1 - probs)
        # compute sum
        res = 0
        for i in range(X.shape[0]):
            res += np.outer(X[i], X[i]) * probs[i]
        return res
    def prob(self, w, X):
        # compute scores
        prob = X @ w
        return 1 / (1 + np.exp(-prob))
```

# 3 - Test our Model on the Toy Datasets :
> In this section, we test our model on the toy datasets with a 10 folds cross validation.

```python
# define our classifier
logreg = LogReg()
for configuration in data:
    X, Y = data[configuration]

    plt.figure(figsize=(15, 5))
    plt.suptitle("Logistic Regression | "+str(configuration), fontsize=16)
    
    # Plot true repartition
    ax = plt.subplot(1, 2, 1)
    ax.set_title("true repartition")
    plt.scatter(X[:, 0], X[:, 1], marker='o', c=Y,s=25, edgecolor='k')
    
    # fit with cross validation
    cv_results = cross_validate(logreg, X, Y, cv=10)
    # fit the data
    logreg.fit(X, Y)
    ax = plt.subplot(1, 2, 2)
    ax.set_title("cross val score on test = "+str(sum(cv_results['test_score'])/len(cv_results['test_score'])))
    # plot the classifier's boundary
    plot_boundary(X, logreg)
```

> The provided implementation of logistic regression provides identical results with sklearn's logistic regression (5 folds cross validation for sklearn's version and 10 folds cross-validation on the from scratch version). A cross validation score of $0.975$ on the classes dataset and $0.88$ on the moons dataset. However, since we trained a binary classification model, we can't classify the blobs dataset.
> * Only two iterations are necessary to achieve the best possible score.

# 4 - Logistic Regression on Medical Data :
> In this section we test logistic regression (both sklearn's version and the version provided above).

```python
logreg_sklearn = linear_model.LogisticRegression(C=1e5)
logreg_scratch = LogReg()
for configuration in medical_data:
    print("------------",configuration,"------------")
    X, Y = medical_data[configuration]
    # sklearn's version
    cv_results = cross_validate(logreg_sklearn, X, Y, cv=10)
    print("---Score obtained with sklearn's version : ",sum(cv_results['test_score'])/len(cv_results['test_score']))
    # hand made
    print("---Coded with different number of parameter updates :")
    for epochs in [1, 4, 8]:
        logreg_scratch = LogReg(epochs = epochs)
        print("Number of iterations : ", epochs)
        cv_results = cross_validate(logreg_scratch, X, Y, cv=10)
        print("Score obtained with the coded version : ",sum(cv_results['test_score'])/len(cv_results['test_score']))
```

> On the **Breast dataset** :
>* Sklearn's version achieves $0.942$ (this dataset is linearly separable.). The coded version achieves a similar (slightly better) score of $0.956$ after only one iteration (for more iteration we don't necessarily observe an improvement, depends on the optimisation step).

> On the **Mice dataset** :
>* Sklean's version achieves $0.995$ (this dataset is linearly separable) : an almost perfect classifier. The coded model doesn't handle multiclass classification.

> As said earlier, a clear limitation to this classifier is this it is linear. We can use kernel methods (RKHS) to deal with this caveat.
> * Project the data onto a new high dimensional space (it can be with an infinite dimension), where the problem is easier (linearly separable).
> * Solve the problem in that space : reason in the high dimensional space, but compute in the original space using the kernel trick (Gramm matrix encompasses all the necessary information)
