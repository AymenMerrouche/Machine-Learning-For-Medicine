# Introdution :
> In this practical session the goal is to predict whether a diabetic patient will resolve or will not resolve his diabetes after a gastric bypass surgery (two classes : **0 for diabestes remission** and **1 for non remission**) based on 4 clinical varaibles : **age of patients** (continuous), **HbA1C** (continuous), **insuline taken** (categorical, yes or not), **other anti-diabetic drugs** are taken (categorical, yes or not). We have data on 200 patients. <br>
We will train two machine learning models (decision trees and random forests) for this binary classification task. Then, we will compare the performances of these models with the state-of-the-art clinical score DiaRem (Diabetes Remission score).

```python
import pandas as pd
import graphviz
import numpy as np
```

# Import the data :

```python
# import the data
data_path = './data/'
data_diabetes = pd.read_table(data_path+"patients_data.txt",sep="\t",header=None)
classes_diabetes = pd.read_table(data_path+"patients_classes.txt",sep="\t",header=None)
```

```python
# clinical variables
data_diabetes
```

```python
# binary classes
classes_diabetes
```

```python
from sklearn.model_selection import train_test_split
# split the data ito a train and test set
X_train, X_test, y_train, y_test = train_test_split\
(data_diabetes, classes_diabetes, test_size=0.33, random_state=42)
```

```python
print("train set length (67% of the data) : ",len(X_train),"\ntest set length (33% of the data) : ",len(X_test))
```

# Decision Trees :

```python
from sklearn import tree
dt_full = tree.DecisionTreeClassifier()
dt_full = dt_full.fit(data_diabetes, classes_diabetes)
```

```python
print("Classification accuracy : ",dt_full.score(data_diabetes, classes_diabetes))
```

> Decicon trees achieve perfect accuracy when trained and tested on the full dataset. This is expected, The hyper-parameter **max_depth** is set to Null, meaning that nodes are expanded until all leaves are pure (of same class) or until all leaves contain less than **min_samples_split** samples which is set to 2. It is obvious that if you increase max_depth, training error will always go down. This combined with the fact that we train and test on the same dataset explains the flawless performance of this model (clear overfitting). 

```python
from IPython.display import display

# save the decision tree into a pdf
feature_names = ['age', 'hba1c', 'insuline taken', 'other drugs taken']
classes = ['DR','NDR']
dot_data = tree.export_graphviz(dt_full, out_file=None,\
                                feature_names=feature_names,class_names=classes,\
                                filled=True, rounded=True,special_characters=True)
graph = graphviz.Source(dot_data)
graph.render("diabetes remission")
```

```python
# display the decision tree
display(graph)
```

> One of the most important requirements when training a machine learning model for a medical task is its interpretability. Understanding the results is just as important as the performance. Decision trees are the go to models when searching for interpretability. <br>
When displaying the constructed decision tree, we clearly understand how a datapoint is classified (rules). We notice that the most discriminating variable is inuline_taken and then age... The nodes give us the corresponding class.


### Grid search (with cross validation) :

> In a traditional ML task fashion, let's tune the hyperparameters of the model using a **grid_search** with a 3 folds cross validation trained and tested on different sets.

```python
from sklearn.model_selection import GridSearchCV
# grid search params
gs_params = {"cv": 3, "n_jobs": -1, "verbose": 1}
```

```python
# parameters for decision trees
dt_params = {"max_depth": [None, 2, 3, 5, 10, 30, 60]} 
grid_dt = GridSearchCV(estimator=tree.DecisionTreeClassifier(), param_grid=dt_params, **gs_params)
grid_dt.fit(X_train, y_train)
print("Best parameters returned by the grid sarch : ", grid_dt.best_params_)
print("Corresponding accuracy on test set : ",grid_dt.score(X_test, y_test))
best_dt = grid_dt.best_estimator_
```

> **Max depth** returned by grid_search is 3 (much smaller than previously, to avoid overfitting of the training data). The corresponding accuracy on the test set is $0.6818$.

```python
# display the corresponding graph
feature_names = ['age', 'hba1c', 'insuline taken', 'other drugs taken']
classes = ['DR','NDR']
dot_data = tree.export_graphviz(best_dt, out_file=None,\
                                feature_names=feature_names,class_names=classes,\
                                filled=True, rounded=True,special_characters=True)
graph = graphviz.Source(dot_data)
display(graph)
```

# Random Forests :

```python
from sklearn.ensemble import RandomForestClassifier
rf_full = RandomForestClassifier(max_depth=2, random_state=0)
rf_full.fit(data_diabetes, classes_diabetes.values.ravel())
```

```python
print("Classification accuracy : ", rf_full.score(data_diabetes, classes_diabetes))
```

> Random forests achieve an accuracy of $0.82$ when trained and tested on the full dataset. It seems odd that decision trees achieve better accuracy than random forests since random forests are bagged trees and we expect them to be able to represent more complex relations between the variables. This is because the maximum depth of the trees used by the model is set to 2 which limits it's complexity. random_state is used to control the randomness and hence get reproducible results.

```python
import matplotlib.pyplot as plt

# feature importances of the forest
features = ['age', 'hba1c', 'insuline_taken', 'other_drugs_taken']
importances = rf_full.feature_importances_
indices = np.argsort(importances)[::-1]

# Plot the impurity-based feature importances of the forest
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

> In an attempt to interpret the model, we display the gini-importance of each feature, the higher this value the more important the feature is. The most important feature is age followed by insuline taken, hba1c and finally other drugs taken. Results are somewhat in line with the best decision tree returned by grid search (the most discriminative variable is age in the corresponding tree).

### Grid search (with cross validation) :

```python
rf_params = {"max_depth": [None, 2, 3, 4, 5, 10, 50],
              "n_estimators": [1, 5, 10, 50, 100, 150]}  # Parameters for random forests
grid_rf = GridSearchCV(estimator=RandomForestClassifier(random_state=0), param_grid=rf_params, **gs_params)
grid_rf.fit(X_train, y_train)
print("Best parameters returned by the grid sarch : ", grid_rf.best_params_)
print("Corresponding accuracy on test set : ",grid_rf.score(X_test, y_test))
best_rf = grid_rf.best_estimator_
```

> The best parameters yielded by cross validation are **max_depth=3** (depth of each tree) and **n_estimators=100** (number of trees to use). The corresponding accuracy is $0.7575$, wich is better than the accuracy achieved but the best decison tree ($0.6818$).

# The DiaRem (Diabetes Remission score) :

> In this section we implement DiaRem, a clinical scoring system designed to predict diabetes remission (DR) after a gastric bypass surgery.

```python
def diarem(data_set):
    """
        Classifies each row in data_set using DiaRem (Diabetes Remission score), varaibles are :
        'age', 'hba1c', 'insuline_taken', 'other_drugs_taken'
    """
    # create a copy of the input dataframe
    df = data_set.copy()
    # renaming columns
    df.columns = ['age', 'hba1c', 'insuline_taken', 'other_drugs_taken']
    # round the values of age and hba1c to meet the darem score-conditions
    df = df.round({'age': 0, 'hba1c': 1})
    # age
    conds = [df.age.lt(40), df.age.ge(40) & df.age.le(49), df.age.ge(50) & df.age.le(59), df.age.ge(60)]
    choices = [0, 1, 2, 3]
    df['diarem_prediction'] = np.select(conds, choices, default=0)
    # hba1c
    conds = [df.hba1c.lt(6.5), df.hba1c.ge(6.5) & df.hba1c.le(6.9),\
             df.hba1c.ge(7) & df.hba1c.le(8.9), df.hba1c.ge(9)]
    choices = [0, 1, 2, 3]
    df['diarem_prediction'] += np.select(conds, choices, default=0)
    # insuline_taken
    df['diarem_prediction'] += 10*df['insuline_taken']
    # other_drugs_taken
    df['diarem_prediction'] += 3*df['other_drugs_taken']
    # predict the class
    df['diarem_prediction'] = np.where(df['diarem_prediction'] < 7, 0, 1)
    return df[['diarem_prediction']]
```

```python
from sklearn.metrics import accuracy_score
prediction_diarem = diarem(data_diabetes)
accuracy_score(classes_diabetes, prediction_diarem)
```

> DiaRem achieves a classification accuracy of $0.765$ (less than the other two methods when trained and tested on the full dataset, however here no overfitting occurs since it's a scoring system)

# Comparison between Decision trees, random forests and DiarRem :

> In this section we compare the predictive performances of decision trees (best model retrieved by cross_validation) VS random forests (best model retrieved by cross_validation) VS DiaRem on a test set using different evaluation metrics (accuracy, f1 score, auc) :

```python
# predictions of each model on test set
dt_y_pred = best_dt.predict(X_test)
rf_y_pred = best_rf.predict(X_test)
diarem_y_pred = diarem(X_test)
```

```python
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score

predictions = {"decision_tree" : dt_y_pred, "random_forest" : rf_y_pred, "DiaRem" : diarem_y_pred}
evaluation_metrics = [accuracy_score, f1_score, roc_auc_score]
for scoring_metric in evaluation_metrics :
    print("---",scoring_metric.__name__,"---")
    for model, y_pred in predictions.items() :
        print("\t",model," : ",scoring_metric(y_test, y_pred))
```

> We compare the best models returned by cross-validation for decision trees and random forests with DiaRem. The tests are performed on a test set unseen by the decision tree model and the random forest model. Across all metrics, random forests achieve best performance followed by DiaRem. The decision tree model achieves poor performance in comparison (not complex enough to capture the underlying relations between the different variables without overfitting the training data). <br>
We can say that DiaRem is still the best choice, in spite of the better predictive performance of random forests :
* DiaRem is simple and clearly interpretable and achieves acceptable performance compared with the other two models.
* Decision Trees are also simple and interpretable, it can be easily converted into rules. However, it achieves poor performance.
* With random forests we loose in interpretability and simplicity. We bagged 100 trees; it is difficult to understand the collective decision of the trees. Feature importance brings us some light on the matter. However, gini importance does not always give an accurate picture of importance https://explained.ai/rf-importance/index.html
